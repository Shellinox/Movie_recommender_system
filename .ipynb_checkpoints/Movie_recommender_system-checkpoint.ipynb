```python
import numpy as np
import pandas as pd
import ast
ast.literal_eval
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies.head()
```

```python
# columns to keep
# genre,id,keywords,title,overview,cast,crew
```

```python
movies=movies[["genres","movie_id","keywords","title","overview","cast","crew"]]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convertGenre(obj):
    genre=[]
    for i in ast.literal_eval(obj):
        genre.append(i["name"])
    return genre
```

```python
movies['genres']=movies['genres'].apply(convertGenre)
```

```python
movies.head()
```

```python
movies.head()
```

```python
def convertCast(obj):
    genre=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            genre.append(i["name"])
            count=count+1
        else:
            break
    return genre
```

```python
movies['cast']=movies["cast"].apply(convertCast)
```

```python
movies.head()
```

```python
def fetchDirector(obj):
    director=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            director.append(i["name"])
            break
    return director
```

```python
movies["crew"]=movies["crew"].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies["overview"]=movies['overview'].apply(lambda str: str.split())
```

```python
movies['cast']=movies['cast'].apply(lambda str: [i.replace(" ","") for i in str])
```

```python
movies['crew']=movies['crew'].apply(lambda str: [i.replace(" ","") for i in str])
movies['keywords']=movies['keywords'].apply(lambda str: [i.replace(" ","") for i in str])
movies['genres']=movies['genres'].apply(lambda str: [i.replace(" ","") for i in str])
```

```python
movies.head()
```

```python
movies["tags"]= movies["cast"]+ movies["crew"]+ movies["keywords"]+ movies["overview"]+ movies["genres"]
```

```python
pre_proc=movies.drop(['keywords',"overview","cast","crew","genres"],axis=1)
```

```python
pre_proc.head(45)
```

```python
pre_proc['tags']=pre_proc['tags'].apply(lambda x: " ".join(x))
```

```python
pre_proc.head()
```

```python
pre_proc["tags"][0]
```

```python
pre_proc['tags']=pre_proc['tags'].apply(lambda x: x.lower())
```

```python
pre_proc["title"]=pre_proc['title'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
pre_proc["tags"]=pre_proc["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(pre_proc["tags"]).toarray()
```

```python
vectors
```

```python
print(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),key=lambda x: x[1],reverse=True)[1:6]
```

```python
def recommedation(movie):
    movie=movie.lower()
    movie_index=pre_proc[pre_proc["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),key=lambda x: x[1],reverse=True)[1:6]
    
    for i in movies_list:
        print(pre_proc.iloc[i[0]].title)
    
```

```python
recommedation("batman begins")
```
